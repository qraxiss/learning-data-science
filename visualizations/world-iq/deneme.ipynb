```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.impute import KNNImputer
```

```python
IQ=pd.read_csv("data.csv", sep=";").drop("Index", axis=1).drop("Country", axis=1)
IQ
```

```python
nan = np.nan
imputer = KNNImputer(n_neighbors=2, weights="uniform")
IQ=pd.DataFrame(imputer.fit_transform(IQ), columns=[IQ.columns[0],IQ.columns[1],IQ.columns[2],IQ.columns[3]])
IQ
```

```python
sns.pairplot(data = IQ, kind="reg")
```

```python
sns.regplot(data=IQ, y = "Earnings Year $", x = "IQ")
```
